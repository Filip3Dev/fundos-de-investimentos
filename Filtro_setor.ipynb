```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
from datetime import date, datetime
from PIL import Image 
import os	
from zipfile import ZipFile
import requests, zipfile, io
from os import chdir, getcwd, listdir
```

```python
nome_pasta = ['FCA','IPE','ITR','DFP']#local onde será salvo
for i in nome_pasta: #para cada pasta dentro da lista
  if not os.path.exists("/content/"+i): #se não existir a pasta  
      os.mkdir("/content/"+i) #crie a pasta 

  caminho = str("/content/"+i) #salve nessa pasta, nesse caminho

dates = pd.date_range('2017', '2022', freq='Y') #datas que serao puxadas
dates #repare q é gerado até o um ano antes

DRI_FCA = pd.DataFrame()
DIVULGACAO_FCA = pd.DataFrame()
ESTRANGEIRO_FCA = pd.DataFrame()
GERAL_FCA = pd.DataFrame()
ESCRITURADOR_FCA = pd.DataFrame()
AUDITOR_FCA = pd.DataFrame()
ANO_FCA = pd.DataFrame()
for i in dates:
  ano = (str(i.to_period('Y'))) #ano 
  url ='http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/FCA/DADOS/fca_cia_aberta_'+ano+'.zip' #link do arquivo
  r = requests.get(url) #baixa o arquivo do link
  z = zipfile.ZipFile(io.BytesIO(r.content))
  z.extractall('/content/FCA')

chdir('/content/FCA') #irá acessar a pasta
for arquivo in listdir(): #para cada arquivo existente dentro da pasta
  aux = (arquivo.upper().split('_'))
  print(aux)
  df = pd.read_csv(arquivo, sep=';', encoding='ISO-8859-1') #le o arquivo
  if  'DRI' in aux:
      DRI_FCA = pd.concat([DRI_FCA,df.copy()])  #Diretor de Relações com Investidores	 - CONTATOS
  elif 'GERAL' in aux: #informações sobre as companhias, setor, atuação etc.
    GERAL_FCA= pd.concat([GERAL_FCA,df.copy()])
  else:
    pass

DRI_FCA.index = DRI_FCA['CNPJ_Companhia'] #cnpj no indice
DRI_FCA = DRI_FCA.drop(columns=['CNPJ_Companhia']) # exclui coluna do cnpj
```

```python
empresas = GERAL_FCA[['Nome_Empresarial','Setor_Atividade','Descricao_Atividade','Situacao_Emissor','Especie_Controle_Acionario']].copy() #critérios de busca
empresas.index = GERAL_FCA['CNPJ_Companhia'] #indice 
empresas = empresas.drop_duplicates() #retira valores duplicados
```
