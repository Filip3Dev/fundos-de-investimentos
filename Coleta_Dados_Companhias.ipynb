#FCA
Criterios de filtragem por características da companhia

```python
DRI_FCA = pd.DataFrame()
DIVULGACAO_FCA = pd.DataFrame()
ACIONISTAS_FCA = pd.DataFrame()
ESTRANGEIRO_FCA = pd.DataFrame()
GERAL_FCA = pd.DataFrame()
ESCRITURADOR_FCA = pd.DataFrame()
AUDITOR_FCA = pd.DataFrame()
ANO_FCA = pd.DataFrame()
for i in dates:
  ano = (str(i.to_period('Y'))) #ano 
  url ='http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/FCA/DADOS/fca_cia_aberta_'+ano+'.zip' #link do arquivo
  r = requests.get(url) #baixa o arquivo do link
  z = zipfile.ZipFile(io.BytesIO(r.content))
  z.extractall('/content/FCA')

chdir('/content/FCA') #irá acessar a pasta
for arquivo in listdir(): #para cada arquivo existente dentro da pasta
  aux = (arquivo.upper().split('_'))
  print(aux)
  df = pd.read_csv(arquivo, sep=';', encoding='ISO-8859-1') #le o arquivo
  if  'DRI' in aux:
      DRI_FCA = pd.concat([DRI_FCA,df.copy()])  # FACULTATIVO - Diretor de Relações com Investidores	 - CONTATOS
  elif 'DIVULGACAO' in aux:
    DIVULGACAO_FCA= pd.concat([DIVULGACAO_FCA,df.copy()])  #FACULTATIVO  - Canal de divulgação de informações 
  elif 'ACIONISTAS' in aux:
    ACIONISTAS_FCA= pd.concat([ACIONISTAS_FCA,df.copy()]) #DESCARTADO - apenas contatos
  elif 'ESTRANGEIRO' in aux: #interessante para saber quais paises investem na empresa
    ESTRANGEIRO_FCA= pd.concat([ESTRANGEIRO_FCA,df.copy()])
  elif 'GERAL' in aux: #iOBRIGATÓRIO - nformações sobre as companhias, setor, atuação etc.
    GERAL_FCA= pd.concat([GERAL_FCA,df.copy()])
  elif 'ESCRITURADOR' in aux: #DESCARTADO - apenas contatos
    ESCRITURADOR_FCA = pd.concat([ESCRITURADOR_FCA,df.copy()])
  elif 'AUDITOR' in aux: #DESCARTADO - apenas contatos
    AUDITOR_FCA = pd.concat([AUDITOR_FCA,df.copy()])
  elif len(aux) == 4:
    ANO_FCA = pd.concat([ANO_FCA,df.copy()])
```
